```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = pd.read_csv(
    r'C:\Users\af.sabirova\Desktop\pyda\HW_10\master.csv')
```

```python
country_dynamics = names.groupby(['country']).sum().sort_values(by='suicides_no', ascending=False).head(10)
country_dynamics.plot(kind='bar', y='suicides_no')
plt.xlabel('country')
plt.ylabel('suicides_no')
plt.title('ТОП СТРАН ПО КОЛИЧЕСТВУ САМОУБИЙСТВ С 1985 ПО 2016 ГГ')
```

```python
#В топ 3 стран по количеству самоубийств на период с 1985 по 2016 гг. - РФ, США и Япония.
#Следующие страны сильно отстают от 3 ТОП лидеров
```

```python
name_generation = names.groupby(['generation']).sum().sort_values(by='suicides_no', ascending=False).head(10)
name_generation.plot(kind='pie', y='suicides_no')
plt.title('РАСПРЕДЕЛЕНИЕ САМОУБЙИСТВ ПО ПОКОЛЕНИЯМ')
```

```python
#Наибольшее количество самоубийств замечено среди поколенич Boomers  (1943-1963 г.р.)
#Cледующим идет поколение Silent (1920—1939 г.р.) и Generation X (1960—1979 г.р.)
```

```python
names_news = names.filter(items = ['year', 'suicides_no'])
year_dynamics = names_news.groupby(['year']).sum().sort_values(by='suicides_no', ascending=False).reset_index ().head(10)
year_dynamics . plot ( kind = 'barh' ,  y = "suicides_no" ,  x = "year" )
plt.title('РАСПРЕДЕЛЕНИЕ САМОУБЙИСТВ ПО ГОДАМ')
```

```python
import seaborn as sns
```

```python
#age_dynamics = names.groupby(['country']).sum().sort_values(by='suicides_no', ascending=False)
country_data = names.groupby(['country-year']).sum().sort_values(by='suicides_no', ascending=False)
country_data.rename(columns={'suicides/100k pop':'suicides_100k_pop', 'HDI for year':'HDI_for_year','gdp_per_capita ($)':'gdp_per_capita'}, inplace=True)
country_data.reset_index()
```

```python
country_data['suicides_100k_pop'] = country_data.suicides_100k_pop.astype('int64')
country_data['gdp_per_capita'] = country_data.gdp_per_capita.astype('int64')
```

```python
sns.distplot(country_data.suicides_100k_pop)
plt.title('РАСПРЕДЕЛЕНИЕ САМОУБИЙСТВ НА 100 ТЫСЯЧ НАСЕЛЕНИЯ ПО СТРАНАМ И ГОДАМ')
```

```python
#На основе графика можно сделать вывод, что на 100 тысяч человек приходится примерно до 150 самоубийств в год. 
```

```python
sns.jointplot(x='suicides_100k_pop', y='gdp_per_capita', 
              data=country_data, kind='scatter')
```

```python
#Из графика можно сделать вывод что показатели ВВП и самоубийства на 100 тысяч человека прямо пропорциональны и при росте ВВП, 
#количество самоубийств будет уменьшаться
```
