```python
#Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг
Результат классификации запишите в столбец class
```

```python
import pandas as pd
data = pd.read_csv(r'C:\Users\af.sabirova\Desktop\pyda\HW_6\ratings.csv')
data.head()
```

```python
def class_search (row):
    if row ['rating'] <= 2:
        return 'низкий рейтинг'
    elif ((row ['rating'] <= 4) & (row ['rating'] > 2)):
        return 'средний рейтинг'
    if ((row ['rating'] ==5) & (row ['rating'] ==4.5)):
        return 'высокий рейтинг'
    return 'не найдено'
data['class'] = data.apply (lambda row: class_search(row), axis=1)
data.head(10)   
```

```python
#Задание 2
Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.
Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
Результат классификации запишите в отдельный столбец region.
```

```python
import pandas as pd
geo = pd.read_csv(r'C:\Users\af.sabirova\Desktop\pyda\7_lek\keywords.csv')
def geo_search(a):
    if 'москва' in a:
        return 'Центр'
    elif 'тула' in a:
        return 'Центр'
    elif 'ярославль' in a:
        return 'Центр'
    elif 'петербург' in a:
        return 'Свеверо-запад'
    elif 'псков' in a:
        return 'Свеверо-запад'
    elif 'мурманск' in a:
        return 'Свеверо-запад'
    elif 'владивосток' in a:
        return 'Дальний Восток'
    elif 'сахалин' in a:
        return 'Дальний Восток'
    elif 'хабаровск' in a :
        return 'Дальний Восток'
    return 'undefined'
geo['region'] = geo['keyword'].apply (geo_search)
geo.head(10)
```

```python
import pandas as pd
import re
data_rating = pd.read_csv(r'C:\Users\af.sabirova\Desktop\pyda\HW_7\ml-latest-small/ratings.csv')
data_movies = pd.read_csv(r'C:\Users\af.sabirova\Desktop\pyda\HW_7\ml-latest-small/movies.csv')
data_new=pd.merge(data_rating, data_movies, on='movieId', how='outer')

data_new['year'] = data_new['title'].str.extract('.*\((.*)\).*', expand=True)

movie_rating = data_new.filter(items = ['year', 'rating'])
movie_rating=movie_rating.groupby('year')['rating'].mean().reset_index()
movie_rating.sort_values(['rating'],ascending=False)    
```
