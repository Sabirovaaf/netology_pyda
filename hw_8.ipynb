```python
1. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId
```

```python
import numpy as np
import pandas as pd
ratings = pd.read_csv(r'C:\Users\af.sabirova\Desktop\pyda\HW_7\ml-latest-small\ratings.csv')
ratings.head()
ratings_grouped = ratings.groupby('userId').count().reset_index()
ratings_filtered = ratings_grouped[ratings_grouped['rating']>=100]
film_fans_user_ids = ratings_filtered['userId'].tolist()
ratings_new=ratings[ ratings['userId'].isin(film_fans_user_ids) ]
rating_final=ratings_new.groupby('userId').agg({'timestamp':[np.max, np.min]})
rating_final.columns=['timestamp_max', 'timestamp_min']
rating_final ['timestamp_delta'] = rating_final.timestamp_max - rating_final.timestamp_min
rating_final.head()

#def groupby_function(data):
#    return ratings.timestamp.max()
#ratings_new.groupby('userId').apply(groupby_function)
```
